```python
#### <font color= green> Function to count the number of parameters in each model
```

```python
def parametersNum():
    NumPar = 0
    for var in tf.trainable_variables():
        shape = var.get_shape()
        varParameter = 1
        for dim in shape:
            varParameter *= dim.value
            NumPar += varParameter
    return(NumPar)
```

#### <font color= green> Now we are going to create __Place Holders__ for input and output (input with 28x28 pixels = 784; output with 10 digits)

```python
x_m1 = tf.placeholder(tf.float32, shape=[None, 784])
x_m2 = tf.placeholder(tf.float32, shape=[None, 784])
y_m1 = tf.placeholder(tf.float32, shape=[None, 10])
y_m2 = tf.placeholder(tf.float32, shape=[None, 10])
y_true_cls_m1 = tf.argmax(y_m1, dimension=1)
y_true_cls_m2 = tf.argmax(y_m2, dimension=1)
keep_prob = tf.placeholder(tf.float32)
```

#### <font color= green> Function to create fully connected layer

```python
def fcLayer(x, num_output, activationFun=None, name=None):

    # define the weights
    weight = tf.get_variable(initializer=tf.random_normal([x.get_shape()[-1].value, num_output], stddev=0.0999), name=name+'/weight')

    # define the bias
    bias = tf.get_variable(shape=[num_output],initializer=tf.constant_initializer(0.1), name=name+'/bias')
    
    # calc sum of neurons
    sum = tf.matmul(x,weight) + bias
        
    # apply activation function if required
    if activationFun:
        sum = activationFun(sum)

    return sum
```

### <font color= blue> In the steps below we want to define different models and their relevant layers

### <font color= red> Define Model 1

#### <font color= green> Define layers

```python
denseLayer1 = fcLayer(x_m1, 784, tf.nn.relu, name='denseLayer1')
dropout1 = tf.nn.dropout(denseLayer1, 0.6)
denseLayer2 = fcLayer(dropout1, 128, tf.nn.relu, name='denseLayer2')
out_m1 = fcLayer(denseLayer2, 10, name='denseLayer3')
y_pred_m1 = tf.nn.softmax(logits=out_m1)
y_pred_cls_m1 = tf.argmax(y_pred_m1, dimension=1)
params_m1 = parametersNum()
print("Number of parameters of Model 1:" , params_m1)
```

#### <font color= green>  Define loss 

```python
cross_entropy_m1 = tf.nn.softmax_cross_entropy_with_logits(labels=y_m1, logits=out_m1)
loss_m1 = tf.reduce_mean(cross_entropy_m1)
```

#### <font color= green> Define optimizer

```python
opt_m1 = tf.train.AdamOptimizer(learning_rate=0.0009)
optimizer_m1 = opt_m1.minimize(loss_m1)
```

#### <font color= green> Defining accuracy

```python
correct_pred_m1 = tf.equal(y_pred_cls_m1, y_true_cls_m1)
accuracy_m1 = tf.reduce_mean(tf.cast(correct_pred_m1, tf.float32))
```

### <font color= red> Define Model 2

#### <font color= green> Define layers

```python
out_m2 = fcLayer(x_m2, 10, tf.nn.relu, name='layer_logit')
y_pred_m2 = tf.nn.softmax(logits=out_m2)
y_pred_cls_m2 = tf.argmax(y_pred_m2, dimension=1)
params_m2 = parametersNum() - params_m1
print(params_m2)
```

#### <font color= green>  Define loss 

```python
cross_entropy_m2 = tf.nn.softmax_cross_entropy_with_logits(labels=y_m2, logits=out_m2)
loss_m2 = tf.reduce_mean(cross_entropy_m2)
```

#### <font color= green> Define optimizer

```python
opt_m2 = tf.train.AdamOptimizer(learning_rate=0.0015)
optimizer_m2 = opt_m2.minimize(loss_m2)
```

#### <font color= green> Defining accuracy

```python
correct_pred_m2 = tf.equal(y_pred_cls_m2, y_true_cls_m2)
accuracy_m2 = tf.reduce_mean(tf.cast(correct_pred_m2, tf.float32))
```

#### <font color= green> Counting Number of Parameters

```python
params_num = 0
for var in tf.trainable_variables():
    shape = var.get_shape()
    varParamsNum = 1
    for dim in shape:
        varParamsNum *= dim.value
    print(varParamsNum)
    params_num += varParamsNum
print(params_num)
```

#### <font color= green> Function for getting weights from the network

```python
def getWeights(layer_name):
    with tf.variable_scope(layer_name, reuse=True):
        var = tf.get_variable('weight')
    return var
```

#### <font color= green> Getting the weights

```python
weights_denseLayer1 = getWeights(layer_name='denseLayer1')
weights_denseLayer2 = getWeights(layer_name='denseLayer2')
weights_out_m1 = getWeights(layer_name='denseLayer3')
weights_out_m2 = getWeights(layer_name='layer_logit')
```

#### <font color= green>Creating a Tensorflow session

```python
sess = tf.Session()
```

#### <font color= green>Initializing the variables

```python
sess.run(tf.global_variables_initializer())
```

#### <font color= green> Getting Gradients

```python
grads_m1 = tf.gradients(loss_m1, weights_out_m1)[0]
print(grads_m1)
grads_m2 = tf.gradients(loss_m2, weights_out_m2)[0]
print(grads_m2)
```

#### <font color= green> Function to calculate grad norm

```python
def calcGradNorm(gradsValues):
    grad_all = 0.0
    for gv in gradsValues:
        grad_all += sum(gv**2)
    
    grad_norm = grad_all ** 0.5
    return grad_norm
```

#### <font color= green> Computing Hessian

```python
hessian_m1 = tf.reduce_sum(tf.hessians(loss_m1, weights_out_m1)[0], axis = 2)
print(hessian_m1)
hessian_m2 = tf.reduce_sum(tf.hessians(loss_m2, weights_out_m2)[0], axis = 2)
print(hessian_m2)
```

#### <font color= green> Training the network

```python
max_epochs=8
batch_size = 64
total_batch_size = int(len(mnist.train.labels) / batch_size)
grad_arr_m1, iteration_arr, train_loss_arr_m1, train_accuracy_arr_m1, weights_out_m1_arr, weights_denseLayer2_arr, weights_denseLayer1_arr, = [], [], [], [], [], [], []

global iteration_ctr
iteration_ctr = 0

for epoch in range(max_epochs):

    print('epoch: {}/{}'.format(epoch+1,max_epochs))

    for batch_ctr in range(total_batch_size):

        # Fetching a batch size of 64 images
        x_batch, y_true_batch = mnist.train.next_batch(batch_size)
        
        # Shuffle the labels before training 
        # np.random.shuffle(y_true_batch)
        
        feed_dict_train_m1 = {x_m1: x_batch, y_m1: y_true_batch}

        # run training for each iteration
        sess.run(optimizer_m1, feed_dict={x_m1: x_batch, y_m1: y_true_batch, keep_prob: 0.5})
        
        # Every 10 steps, compute the loss and accuracy of training
        if batch_ctr % 10 == 0:
            train_loss_m1, train_accuracy_m1 = sess.run([loss_m1, accuracy_m1], feed_dict=feed_dict_train_m1)
            train_accuracy_arr_m1.append(train_accuracy_m1)
            train_loss_arr_m1.append(train_loss_m1)
            iteration_ctr +=1
            iteration_arr.append(iteration_ctr)
            
            # Computing gradient and hessian
            grads_vals_m1, hess_vals_m1 = sess.run([grads_m1, hessian_m1], feed_dict=feed_dict_train_m1)

            # Calculate gradient norm
            grad_norm_m1 = calcGradNorm(grads_vals_m1)
            grad_arr_m1.append(grad_norm_m1)
            
            # Every 100 steps, print the loss and accuracy of training
            if batch_ctr % 100 == 0:
                print('Iteration_m1 {:>6}   Training Loss {:>1.4}  Training Accuracy {:>6.1%}'.format(batch_ctr+1, train_loss_m1, train_accuracy_m1))

        # after every 3 epochs collect weights
        if epoch % 3 == 0:
      
            # Model 1
            weights_denseLayer1_tensor = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope='denseLayer1/weight:0')
            weights_denseLayer1_values = sess.run(weights_denseLayer1_tensor, feed_dict=feed_dict_train_m1)
            weights_denseLayer1_arr.append(weights_denseLayer1_values)
            
            weights_denseLayer2_tensor = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope='denseLayer2/weight:0')
            weights_denseLayer2_values = sess.run(weights_denseLayer2_tensor, feed_dict=feed_dict_train_m1)
            weights_denseLayer2_arr.append(weights_denseLayer2_values)

            weights_out_m1_tensor = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope='denseLayer3/weight:0')
            weights_out_m1_values = sess.run(weights_out_m1_tensor, feed_dict=feed_dict_train_m1)
            weights_out_m1_arr.append(weights_out_m1_values)

print('Iteration_m1 {:>6} Training Loss {:>1.6} Training Accuracy {:>6.1%}'.format(batch_ctr+1, train_loss_m1, train_accuracy_m1))
```

```python
max_epochs=8
batch_size = 64
total_batch_size = int(len(mnist.train.labels) / batch_size)
grad_arr_m2, iteration_arr, train_loss_arr_m2, train_accuracy_arr_m2, weights_out_m2_arr = [], [], [], [], []

global iteration_ctr
iteration_ctr = 0

for epoch in range(max_epochs):

    print('epoch: {}/{}'.format(epoch+1,max_epochs))

    for batch_ctr in range(total_batch_size):

        # Fetching a batch size of 64 images
        x_batch, y_true_batch = mnist.train.next_batch(batch_size)
        
        # Shuffle the labels before training 
        # np.random.shuffle(y_true_batch)
        
        feed_dict_train_m2 = {x_m2: x_batch, y_m2: y_true_batch}

        # run training for each iteration
        sess.run(optimizer_m2, feed_dict={x_m2: x_batch, y_m2: y_true_batch, keep_prob: 0.5})
        
        # Every 10 steps, compute the loss and accuracy of training
        if batch_ctr % 10 == 0:
           
            iteration_ctr +=1
            iteration_arr.append(iteration_ctr)         
            train_loss_m2, train_accuracy_m2 = sess.run([loss_m2, accuracy_m2], feed_dict=feed_dict_train_m2)
            train_accuracy_arr_m2.append(train_accuracy_m2)
            train_loss_arr_m2.append(train_loss_m2)
            
            # Computing gradient and hessian
            grads_vals_m2, hess_vals_m2 = sess.run([grads_m2, hessian_m2], feed_dict=feed_dict_train_m2)

            # Calculate gradient norm
            grad_norm_m2 = calcGradNorm(grads_vals_m2)
            grad_arr_m2.append(grad_norm_m2)
            
            # Every 100 steps, print the loss and accuracy of training
            if batch_ctr % 100 == 0:
                print('Iteration_m2 {:>6}   Training Loss {:>1.4}  Training Accuracy {:>6.1%}'.format(batch_ctr+1, train_loss_m2, train_accuracy_m2))

        # after every 3 epochs collect weights
        if epoch % 3 == 0:

            # Model 2
            weights_out_m2_tensor = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope='layer_logit/weight:0')
            weights_out_m2_values = sess.run(weights_out_m2_tensor, feed_dict=feed_dict_train_m2)
            weights_out_m2_arr.append(weights_out_m2_values)

print('Iteration_m2 {:>6} Training Loss {:>1.6} Training Accuracy {:>6.1%}'.format(batch_ctr+1, train_loss_m2, train_accuracy_m2))
```

#### <font color= green>Testing accuracy on test data

```python
mnist.test.cls = np.argmax(mnist.test.labels, axis=1)
test_batch_size = 256
numTestSamples = len(mnist.test.images)
y_pred_m1 = np.zeros(shape=numTestSamples, dtype=np.int) # place holder for classes
y_pred_m2 = np.zeros(shape=numTestSamples, dtype=np.int) # place holder for classes

i = 0
while i < numTestSamples:
    j = min(i + test_batch_size, numTestSamples) # denotes the ending index for the next batch

    # Get the images between index i and j
    img = mnist.test.images[i:j, :]

    # Get the associated labels
    labels = mnist.test.labels[i:j, :]

    # Create feed-dict
    feed_dict_m1 = {x_m1: img, y_m1: labels}
    feed_dict_m2 = {x_m2: img, y_m2: labels}

    # predict classes
    y_pred_m1[i:j] = sess.run(y_pred_cls_m1, feed_dict=feed_dict_m1)
    y_pred_m2[i:j] = sess.run(y_pred_cls_m2, feed_dict=feed_dict_m2)

    i = j

# calc accuracy
y_true = mnist.test.cls
correct_m1 = (y_true == y_pred_m1)
correct_m2 = (y_true == y_pred_m2)

numCorrect_m1 = correct_m1.sum()
numCorrect_m2 = correct_m2.sum()

acc_m1 = float(numCorrect_m1) / numTestSamples
acc_m2 = float(numCorrect_m2) / numTestSamples

print("Model 1 accuracy on Test-Set: {0:.2%} ({1} / {2})".format(acc_m1, numCorrect_m1, numTestSamples))
print("Model 2 accuracy on Test-Set: {0:.2%} ({1} / {2})".format(acc_m2, numCorrect_m2, numTestSamples))
```

Principal component analysis (PCA)

```python
def plotWeights(weights):
    # Retrieving the values of the weight-variables from TensorFlow.
    w = []
    for x in weights:
        w_var = sess.run(x)
        w.append(w_var)
    
    pca = PCA(n_components=2)
    
    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel('Principal Component 1', fontsize = 15)
    ax.set_ylabel('Principal Component 2', fontsize = 15)
    
    for i in w:
        principalComponents = pca.fit_transform(i)
        ax.scatter(principalComponents[:,0], principalComponents[:,1], label=i.shape, alpha=0.5)

    ax.legend()
    plt.show()
```

```python
plotWeights(weights=[weights_denseLayer2,  weights_out_m1])
plotWeights(weights=[weights_denseLayer1, weights_denseLayer2, weights_out_m1])
```
